```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.nn.modules.module import Module
from torch.nn.parameter import Parameter 
import numpy as np
from sklearn.manifold import TSNE
```

```python
# Function for visualization
import torch
import networkx as nx
import matplotlib.pyplot as plt


def visualize(h, color, epoch=None, loss=None):
    plt.figure(figsize=(7,7))
    plt.xticks([])
    plt.yticks([])

    if torch.is_tensor(h):
        h = h.detach().cpu().numpy()
        plt.scatter(h[:, 0], h[:, 1], s=140, c=color, cmap="Set2")
        if epoch is not None and loss is not None:
            plt.xlabel(f'Epoch: {epoch}, Loss: {loss.item():.4f}', fontsize=16)
    else:
        nx.draw_networkx(G, pos=nx.spring_layout(G, seed=42), with_labels=False,
                         node_color=color, cmap="Set2")
    plt.show()
```

```python
# Use KarateClub as a test case
from torch_geometric.datasets import KarateClub
dataset = KarateClub()
data = dataset[0]

from torch_geometric.utils import to_networkx
G = to_networkx(data, to_undirected = True)
visualize(G, color = data.y)
```

```python
# Model building, it is a vanilla version of Graph convolution neural
# network described as in https://tkipf.github.io/graph-convolutional-networks/

# in this vanilla version, we just constructed a simple 2 hidden layer
# GCNN. The input to the last layer is the embedding.
class GCNN(Module):
    def __init__(self, infeatures, outfeatures, \
                 hidden_dim, dropout = 0.0, act = F.relu):
        super(GCNN, self).__init__()
        self.infeatures = infeatures
        self.outfeatures = outfeatures
        self.hidden_dim = hidden_dim
        self.dropout = dropout
        self.act = act
        self.weight1 = Parameter(torch.FloatTensor(infeatures, hidden_dim))
        self.weight2 = Parameter(torch.FloatTensor(hidden_dim, outfeatures))
        self.classifier = nn.Linear(outfeatures, 4)
        self.reset_parameters()
        
    def reset_parameters(self):
        torch.nn.init.xavier_normal_(self.weight1)
        torch.nn.init.xavier_normal_(self.weight2)
        
    def forward(self, input_, adj):
        support = torch.mm(input_, self.weight1)
        support = torch.spmm(adj, support)
        support = self.act(support)
        
        support = torch.mm(support, self.weight2)
        support = torch.spmm(adj,support)
        embedding = self.act(support)
        output = self.classifier(embedding)
        return output, embedding
```

```python
model = GCNN(dataset.num_features, 2, hidden_dim = 10)
adj_mat = np.zeros((34,34))
for i,j in data.edge_index.reshape((-1,2)):
    adj_mat[i,j] = 1
    adj_mat[j,i] = 1
    
adj_mat += np.eye(34)
D_inv = np.linalg.inv(np.diag(adj_mat.sum(axis = 0)))

adj_mat = np.matmul(D_inv**(0.5), adj_mat.dot(D_inv**(0.5)))
adj_mat = torch.FloatTensor(adj_mat)

class_infos, embed = model(data.x, adj_mat) 

temp = embed.detach().numpy()
colors = ['blue','green','orange','grey']
color_pre_node = [colors[label] for label in data.y]
plt.figure(figsize=(7,7))
plt.scatter(temp[:, 0], temp[:, 1], linewidth=7, c = color_pre_node)
```

Due to the naivety of this model and the dataset, we use the full dataset and the labels for training and did not use the stochastic gradient optimization.

```python
loss_func = torch.nn.CrossEntropyLoss()
optim = torch.optim.Adam(model.parameters(), lr = 0.02)
loss_his = []

def train(data):
    optim.zero_grad()
    output, embedding = model(data.x, adj_mat)
    loss = loss_func(output, data.y)
    loss.backward()
    optim.step()
    return loss, embedding

colors = ['blue','green','orange','grey']
color_pre_node = [colors[label] for label in data.y]
for epoch in range(1000):
    loss, embedding = train(data)
    loss_his.append(loss)
    if epoch % 50 == 0:
        visualize(embedding, color = data.y, epoch = epoch, loss = loss)
```
